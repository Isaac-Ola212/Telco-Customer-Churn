## Statistical Foundations

Understanding basic statistical concepts is important for analysing datasets and building machine learning models. The following concepts support the analysis performed in this project.

**Mean** - The mean (average) is the sum of all values divided by the number of observations.

Formula:

Mean = (x₁ + x₂ + x₃ + ... + xₙ) / n

In this project, the mean helps understand the average monthly charges or tenure of customers in the dataset.

 **Median** - The median is the middle value in a dataset when the values are arranged in ascending order.

* If the number of observations is odd, the median is the middle value.
* If it is even, the median is the average of the two middle values.

The median is useful because it reduces the impact of outliers compared to the mean.

Example in this dataset: Median MonthlyCharges helps identify the typical cost paid by customers.

**Standard Deviation** - The standard deviation measures how spread out the values in a dataset are around the mean.

Formula:

σ = √( Σ (x - μ)² / N )

Where:
x = data value
μ = mean
N = number of observations

A high standard deviation means the values vary widely, while a low standard deviation means the values are close to the mean.

In this project, it helps measure variation in:
Monthly charges
Total charges
Customer tenure

## Hypothesis Testing

Hypothesis testing is used to determine whether a statistical assumption about a dataset is likely to be true.

It involves two hypotheses:

**Null Hypothesis (H₀)**
There is no relationship between variables.

**Alternative Hypothesis (H₁)**
There is a significant relationship between variables.

Example in churn analysis, hypothesis can claim that:
* H₀: Contract type has no effect on customer churn.
* H₁: Contract type affects customer churn.

Through data analysis and visualisation, we evaluate whether patterns suggest rejecting the null hypothesis. 


Hypothesis testing helps determine whether patterns observed in the data are meaningful or simply due to random chance.

1. **P-Value**

A p-value is a statistical measure used to determine the significance of results. It ranges between 0 and 1.

Typical interpretation:
* p < 0.05 → statistically significant, reject the null hypothesis
* p < 0.01 → highly significant
* p > 0.05 → not statistically significant

In churn analysis, p-values help determine whether variables such as contract type or monthly charges significantly influence churn.

2. **Significance Level**

The significance level (α) is the threshold used to decide whether a result is statistically significant. It is typically set at 0.05 (5%).

If:
* p-value < significance level, the null hypothesis is rejected.
* p-value > significance level, the null hypothesis cannot be rejected.

##### Hypothesis Tests Used in Data Analysis

Several statistical tests can be used to validate hypotheses.

* **Welch’s T-Test** 

A Welch’s T-test is used to compare the means of two groups to determine whether they are statistically different.

Unlike a standard t-test, the Welch test does not assume equal variance or equal sample sizes, making it more flexible.
For example, it could be used to compare:
Average monthly charges of churned vs non-churned customers

* **Chi-Square Test**

The Chi-Square test is used to analyse categorical data and determine whether there is a relationship between variables.

In this project it could be used to evaluate relationships such as:
- Contract type vs churn
- Internet service vs churn
The test compares observed values with expected values if the null hypothesis were true.

* **ANOVA (Analysis of Variance)**

ANOVA is used to compare the means of three or more groups to determine if there are statistically significant differences between them.

For example, ANOVA could be used to compare:
Average monthly charges across different contract types

The purpose is to determine whether differences are real or simply due to random chance.
