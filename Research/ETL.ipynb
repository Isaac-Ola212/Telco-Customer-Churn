# Section 1

#### EXTRACT

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load the Raw Dataset

df = pd.read_csv("dataset/raw/Telco-Customer-Churn.csv")
df.head()
```

---

# Section 2

#### TRANSFORM

```python
# Check the structure of the dataset
df.info()
```

```python
# Change TotalCharges Data Type
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Handle Missing Values
df.dropna(inplace=True)

# Confirm no missing values
df.isnull().sum()
```

```python
# Check and Remove Duplicates 
df.duplicated().sum()
```

```python
# Convert Binary Columns to Numerical
binary_cols = [
    "Partner",
    "Dependents",
    "PhoneService",
    "PaperlessBilling",
    "Churn"
]
df[binary_cols] = df[binary_cols].replace({"Yes": 1, "No": 0})

df[binary_cols].head()
```

```python
# Feature Engineering
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 60, 100],
    labels=["0-1yr", "1-2yr", "2-4yr", "4-5yr", "5+yr"]
)

df[["tenure", "tenure_group"]].head(10)
```

```python
# Final Data Check
df.info()
df.describe()
df.head()
```

# Section 3

#### LOAD

```python
# Save cleaned dataset
df.to_csv("dataset/processed/Telco-Customer-Churn-Cleaned.csv", index=False)

print("Cleaned dataset saved successfully.")
```
